```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
tesla_data = yf.download('TSLA', start='2021-01-01', end='2021-12-31')
tesla_data = tesla_data.reset_index()
print("First five rows of Tesla stock data:")
print(tesla_data.head())
```

```python

import matplotlib.pyplot as plt

def make_graph():
    tesla_data = yf.download('TSLA', start='2021-01-01', end='2021-12-31')
    
    plt.figure(figsize=(10, 6))
    plt.plot(tesla_data.index, tesla_data['Close'], marker='o', linestyle='-', color='b', label='Tesla Stock Price')
    
    plt.title('Tesla Stock Price (2021)')
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    
    plt.grid(True)
    plt.legend()
    plt.show()
make_graph()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

def make_graph():
    gme_data = yf.download('GME', start='2021-01-01', end='2021-12-31')
    
    plt.figure(figsize=(10, 6))
    plt.plot(gme_data.index, gme_data['Close'], marker='o', linestyle='-', color='r', label='GameStop Stock Price')
    
    plt.title('GameStop Stock Price (2021)')
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    
    plt.grid(True)
    plt.legend()
    
    plt.show()

make_graph()
```
